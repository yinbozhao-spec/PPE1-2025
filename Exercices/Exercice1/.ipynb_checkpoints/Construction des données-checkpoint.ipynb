```python
from jiwer import wer, cer

reference = "Le lendemain matin, Catherine peigna les grands cheveux noirs de sa petite fille avec soin"
hypothese = "Le lndemain matin, Catherino peigna les grands cheveux noirs des sn petite fillo avec soin"

cer = cer(reference, hypothese)
wer = wer(reference, hypothese)

print(f"CER : {cer}")
print(f"WER : {wer}")
```
